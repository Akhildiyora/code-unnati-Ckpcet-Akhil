# Random Number generation using numpy

```python
from numpy import random 
```

```python
x = random.randint(100)
print(x)
```

```python
a = random.randint(1000)
print(a)
b = random.randint(1000)
print(b)
c = random.randint(1000)
print(c)
d = random.randint(1000)
print(d)
e = random.randint(1000)
print(e)
```

```python
y = random.randint(100)
while True:
    print(y)
```

```python
for i in range(5):
    print(random.randint(1000))
```

```python
#Generate random float

x = random.rand()
print(x)
```

```python
for i in range(10):
    x = random.rand()
print(x)
```

```python
#Generate random Array

from numpy import random
x = random.randint(100,size=(5,8))
print(x)
```

```python
x.shape
```

# measure center of tendency

```python
#1
print(x)
#min function
print(np.amin(x))
#max function
print(np.amax(x))
```

```python
#2  finding mean, median, standerd Deviation and Veriance

np.mean(x)
```

```python
np.median(x)
```

```python
np.std(x)
```

```python
np.var(x)
```
