# UNIT 3 Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Graph

```python
#x axis valus
x=[1,2,4,6,8]
#y axis  valus
y=[7,9,11,13,18]
```

```python
# naming the x axis
plt.xlabel('sales')
# naming the y axis
plt.ylabel('price')
# to give a  title to graph
plt.title('my first graph')
# Plotting the graph using the points
plt.plot(x,y)
```

```python
# for dotted line
plt.plot(x,y,linestyle='dotted')
```

```python
# change colour of line
plt.plot(x,y,color='r')
```

```python
# To make big pointer in point AND using ms chanf size
plt.plot(x,y,ms=10,marker ='o')#(O,#,+,etc)
```

# bar

```python
#x axis valus
x=[1,2,4,6,8]
#y axis  valus
y=[7,9,11,13,18]

plt.bar(x,y)
```

```python
plt.bar(x,y,color='r')
plt.plot(x,y,marker='+',color='g',ms=25)
```
